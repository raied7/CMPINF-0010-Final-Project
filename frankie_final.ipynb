# ARRESTS
My metric is based off Pittsburgh Police arrests over the past couple years. 
This will show how safe/dangerous these neighborhoods are.  Link: https://data.wprdc.org/dataset/arrest-data

```python
# Import pandas and dataset
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt


arrests = pd.read_csv("https://data.wprdc.org/datastore/dump/e03a89dd-134a-4ee8-a2bd-62c40aeebc6f")
                            
```

## Show the dataset neighborhoods

```python

neighborhoods = arrests["INCIDENTNEIGHBORHOOD"].unique()
print(neighborhoods)
print(str(len(neighborhoods))+" neighborhoods")
```

```python
# Remove unneeded columns 
better = arrests[['INCIDENTNEIGHBORHOOD','OFFENSES']]
```

# Graph PieChart of arrests

```python
arrests["INCIDENTNEIGHBORHOOD"].value_counts().plot.pie(y='OFFENSES', subplots=False, figsize=(50,50))
```

# Show most dangerous neighborhoods

```python
number = better.groupby('INCIDENTNEIGHBORHOOD').count()
number.sort_values('OFFENSES', ascending=False).head(20)
```

# Most Common Offenses

```python
offenses = arrests["OFFENSES"].unique()

arrests["OFFENSES"].value_counts().head(10)
```

```python
# Find the average number of arrests
number.mean()
```

# Show safest neighborhoods

```python
number.sort_values('OFFENSES', ascending=False).tail(20)
```

# Plot Data

```python

safest_hoods=number.sort_values(by=['OFFENSES'], ascending=True).head(30)
safest_hoods.plot.bar()
plt.xlabel("NEIGHBORHOODS", size = 15)


```
